Creating a new column for the regions where we recive most of the sunlight

```python
conditions = [
    df['Sunlight_Hours'] < 5,
    (df['Sunlight_Hours'] >= 5) & (df['Sunlight_Hours'] < 8),
    (df['Sunlight_Hours'] >= 8) & (df['Sunlight_Hours'] < 9),
    df['Sunlight_Hours'] > 9  # Condition for South India
]

choices = ['North India', 'West India', 'East India', 'South India']

df['Region'] = np.select(conditions, choices, default=np.nan)
```

DAILY

```python
import random

#  Function to generate random values for clay soil moisture (30-40)
def generate_clay_moisture():
    return np.random.uniform(30, 40)

# Function to generate random values for loam soil moisture (25-35)
def generate_loam_moisture():
    return np.random.uniform(25, 35)
    pass

# Function to generate random values for sandy soil moisture (25-35)
def generate_sandy_moisture():
    return np.random.uniform(15, 30)
    pass

conditions = [
    (df['Soil_Type'] == 'clay') & (df['Water_Frequency'] == 'daily'),
    (df['Soil_Type'] == 'loam') & (df['Water_Frequency'] == 'daily'),
    (df['Soil_Type'] == 'sandy') & (df['Water_Frequency'] == 'daily')
]

choices = [
    generate_clay_moisture(),
    generate_loam_moisture(),
    generate_sandy_moisture()
]

df['Soil_moisture_data'] = np.select(conditions, choices, default=df['Soil_moisture_data'])
```

```python
#  Function to generate random values for clay soil moisture (30-40)
def generate_clay_moisture():
    return np.random.uniform(12, 28)

# Function to generate random values for loam soil moisture (25-35)
def generate_loam_moisture():
    return np.random.uniform(10, 25)
    pass

# Function to generate random values for sandy soil moisture (25-35)
def generate_sandy_moisture():
    return np.random.uniform(5, 15)
    pass

conditions = [
    (df['Soil_Type'] == 'clay') & (df['Water_Frequency'] == 'weekly'),
    (df['Soil_Type'] == 'loam') & (df['Water_Frequency'] == 'weekly'),
    (df['Soil_Type'] == 'sandy') & (df['Water_Frequency'] == 'weekly')
]

choices = [
    generate_clay_moisture(),
    generate_loam_moisture(),
    generate_sandy_moisture()
]

df['Soil_moisture_data'] = np.select(conditions, choices, default=df['Soil_moisture_data'])
```
